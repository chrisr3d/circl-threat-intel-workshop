```python
import pymisp
import urllib3
import getpass

MISP_BASEURL = "https://training.misp-community.org"
MISP_API_KEY = getpass.getpass("Enter your MISP AuthKey:")

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

misp = pymisp.PyMISP(
    MISP_BASEURL,
    MISP_API_KEY,
    ssl=False # Disable SSL verification
)
print(f"Connected to MISP {misp.root_url} running version: {misp.version['version']}")
```

```python
from urllib.parse import urlparse

unique_domains = set()

urls = misp.search("attributes", type_attribute=["url"], publish_timestamp="30d", to_ids=True, published=True, pythonify=True)
for url in urls:
    parsed = urlparse(url.value)
    domain = parsed.hostname
    if domain and not all(c.isdigit() or c == '.' for c in domain):
        unique_domains.add(domain)

domains = misp.search("attributes", type_attribute=["domain"], publish_timestamp="30d", to_ids=True, published=True, pythonify=True)
for domain in domains:
    unique_domains.add(domain.value)

print(f"\n{len(unique_domains)} Unique Domains:")
for domain in sorted(unique_domains):
    print(domain)
```
